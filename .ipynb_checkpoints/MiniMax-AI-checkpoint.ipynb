# MiniMaxAI - Generation 2

Die zweite Generation der Schach-KI nutzt den implementierten Minimax-Algorithmus fr die Berechnung des bestmglichen Zuges. Ihr wird die Suchtiefe als Argument bergeben.

```julia
struct MiniMaxAI 
    depth::Int64
end
```

Um dieser Anforderung gerecht zu werden, definieren wir die folgende Funktion `getNextMove`:
<br>

**Input:**
+ ai &rarr; die MiniMaxAI
+ extboard &rarr; das Spielbrett
+ _... &rarr; ein Platzhalter, der es ermglicht weitere Parameter mitzugeben 

**Output:**
+ move &rarr; ein Zug, zufllig gewhlt aus allen bestmglichen Zgen


```julia
function getNextMove(ai::MiniMaxAI, extboard::ExtendedBoard, _...)::Chess.Move
    score, moves = minimax(extboard, ai.depth)
    return rand(moves)
end
```
