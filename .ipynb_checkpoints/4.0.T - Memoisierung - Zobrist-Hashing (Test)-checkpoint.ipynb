Der folgenden Tests validieren, dass eine Transposition stets den gleichen Hash erzeugt.

```julia
extboard1 = ExtendedBoard(Chess.startboard())
extboard2 = deepcopy(extboard1)

domove!(extboard1, Chess.movefromstring("e2e4"))
domove!(extboard1, Chess.movefromstring("e7e5"))
domove!(extboard1, Chess.movefromstring("d2d4"))
domove!(extboard1, Chess.movefromstring("e5d4"))

domove!(extboard2, Chess.movefromstring("d2d4"))
domove!(extboard2, Chess.movefromstring("e7e5"))
domove!(extboard2, Chess.movefromstring("e2e4"))
domove!(extboard2, Chess.movefromstring("e5d4"))
                                
IJulia.display(HTML(IJulia.html(extboard1.board)))
                                        
@assert extboard1.zobrist.hash == extboard2.zobrist.hash

domove!(extboard1, Chess.movefromstring("d1h5"))
domove!(extboard1, Chess.movefromstring("g7f6"))
domove!(extboard1, Chess.movefromstring("h5g4"))
domove!(extboard1, Chess.movefromstring("f7f5"))
                
domove!(extboard2, Chess.movefromstring("d1f3"))
domove!(extboard2, Chess.movefromstring("g7f6"))
domove!(extboard2, Chess.movefromstring("f3g4"))
domove!(extboard2, Chess.movefromstring("f7f5"))
                
IJulia.display(HTML(IJulia.html(extboard1.board)))
                
@assert extboard1.zobrist.hash == extboard2.zobrist.hash

domove!(extboard1, Chess.movefromstring("f1e2"))
domove!(extboard1, Chess.movefromstring("a7a5"))
domove!(extboard1, Chess.movefromstring("g1f3"))
domove!(extboard1, Chess.movefromstring("f5g4"))
domove!(extboard1, Chess.movefromstring("e1g1"))

domove!(extboard2, Chess.movefromstring("g1f3"))
domove!(extboard2, Chess.movefromstring("f5g4"))
domove!(extboard2, Chess.movefromstring("f1e2"))
domove!(extboard2, Chess.movefromstring("a7a5"))
domove!(extboard2, Chess.movefromstring("e1g1"))
                                        
IJulia.display(HTML(IJulia.html(extboard1.board)))
                                        
@assert extboard1.zobrist.hash == extboard2.zobrist.hash
```
